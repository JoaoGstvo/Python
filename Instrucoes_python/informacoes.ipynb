```python
# Listas

# Criar uma lista
minha_lista = [1, 2, 3, 4]

# # Adicionar elementos
minha_lista.append(5)         # Adiciona ao final
minha_lista.insert(1, 10)     # Adiciona na posição específica

# # Remover elementos
minha_lista.remove(10)        # Remove o primeiro elemento encontrado
minha_lista.pop(2)            # Remove o elemento no índice 2
minha_lista.pop()             # Remove o último elemento
minha_lista.clear()           # Remove todos os elementos

# # Acessar elementos
valor = minha_lista[0]        # Acessa o primeiro elemento
sub_lista = minha_lista[1:3]  # Acessa uma sublista (do índice 1 ao 2)

# Verificar presença de um item
if 3 in minha_lista:          
    print("O número 3 está na lista")

# Tamanho da lista
tamanho = len(minha_lista)

# Ordenar uma lista
minha_lista.sort()                    # Ordena em ordem crescente
minha_lista.sort(reverse=True)        # Ordena em ordem decrescente
lista_ordenada = sorted(minha_lista)  # Retorna uma nova lista ordenada

# Reverter a ordem da lista
minha_lista.reverse()

# Contar ocorrências de um item
ocorrencias = minha_lista.count(2)

# Encontrar o índice de um item
indice = minha_lista.index(3)

# Concatenar listas
nova_lista = minha_lista + [6, 7, 8]

# Repetir uma lista
lista_repetida = minha_lista * 2

# Copiar uma lista
copia_lista = minha_lista.copy()

# Compreensão de lista (List Comprehension)
quadrados = [x**2 for x in minha_lista]
```

```python
# Conjuntos

# Criar um conjunto (set)
meu_conjunto = {1, 2, 3, 4}

# Adicionar elementos
meu_conjunto.add(5)               # Adiciona o elemento 5

# Remover elementos
meu_conjunto.remove(2)            # Remove o elemento 2 (gera erro se não existir)
meu_conjunto.discard(3)           # Remove o elemento 3 (não gera erro se não existir)
meu_conjunto.pop()                # Remove e retorna um elemento arbitrário
meu_conjunto.clear()              # Remove todos os elementos

# Verificar a presença de um item
if 4 in meu_conjunto:
    print("O número 4 está no conjunto")

# Tamanho do conjunto
tamanho = len(meu_conjunto)

# União de dois conjuntos
outro_conjunto = {3, 4, 5}
uniao = meu_conjunto.union(outro_conjunto)  # Retorna a união dos conjuntos

# Interseção de dois conjuntos
intersecao = meu_conjunto.intersection(outro_conjunto)  # Retorna os elementos comuns

# Diferença entre dois conjuntos
diferenca = meu_conjunto.difference(outro_conjunto)  # Retorna os elementos que estão no meu_conjunto, mas não no outro_conjunto

# Diferença simétrica entre dois conjuntos
dif_simetrica = meu_conjunto.symmetric_difference(outro_conjunto)  # Retorna os elementos que estão em um conjunto ou no outro, mas não em ambos

# Atualizar o conjunto com outro conjunto
meu_conjunto.update(outro_conjunto)  # Adiciona os elementos de outro_conjunto ao meu_conjunto

# Atualizar com a interseção
meu_conjunto.intersection_update(outro_conjunto)  # Mantém apenas os elementos presentes em ambos os conjuntos

# Atualizar com a diferença
meu_conjunto.difference_update(outro_conjunto)  # Remove os elementos de outro_conjunto do meu_conjunto

# Atualizar com a diferença simétrica
meu_conjunto.symmetric_difference_update(outro_conjunto)  # Mantém apenas os elementos que estão em um ou outro, mas não em ambos

# Subconjunto e superconjunto
e_subconjunto = meu_conjunto.issubset(outro_conjunto)         # Verifica se meu_conjunto é subconjunto de outro_conjunto
e_superconjunto = meu_conjunto.issuperset(outro_conjunto)     # Verifica se meu_conjunto é superconjunto de outro_conjunto

# Conjuntos disjuntos
e_disjunto = meu_conjunto.isdisjoint(outro_conjunto)     # Verifica se os conjuntos não têm elementos em comum

# Copiar um conjunto
copia_conjunto = meu_conjunto.copy()
```
